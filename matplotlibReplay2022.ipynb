```python
a=range(0,7)
b=[55,21,61,98,85,52,99]
plt.bar(a,b)
plt.show()
```

```python
m=[str(i) for i in range(1,13)]
y_pos=np.arange(len(m))
sales=[20*i for i in range(1,13)]
plt.bar(y_pos,sales,alpha=0.5)
plt.ylabel('urage')
plt.title('dales trends')
plt.show()
```

```python
import random
x=[random.randint(1,100) for i in range(1,7)]
y=[random.randint(1,100) for i in range(1,7)]
plt.scatter(x,y)
plt.show()
```

```python
labels=['A','B','C','D','E']
data=[54,32,18,44,29]
ex=[0.1,0,0,0,0]
plt.pie(data,explode=ex,labels=labels,autopct='%1.1f%%',counterclock=False)
plt.show()
```

```python
labels=['Python', 'Ruby','Java','C++','PHP']
size=[40,12,24,16,9]
colors=['navy','yellow','blue','gold','tomato']
plt.pie(size,labels=labels,colors=colors)
plt.axis('equal')
plt.show()
```

```python
x=np.random.normal(10,10,1000)
plt.hist(x)
plt.show()
```

```python
math=[162,168,172,181,176,168,173,175,162,169]
x=np.array(math)
plt.title('height')
plt.grid()
plt.boxplot(x)
plt.show()
```

```python
math=[82,75,50,73,65,95,78,93,71,83]
english=[77,92,62,77,64,45,28,60,37,86]
points=(math,english)
fig,ax=plt.subplots()
bp=ax.boxplot(points)
ax.set_xticklabels(['math','english'])
plt.title('exam')
plt.grid()
plt.show()
```

```python
x1,y1=range(0,5),[10,41,44,29,85]
x2,y2=range(0,5),[59,55,77,15,47]
fig=plt.figure()
a1=fig.add_subplot(1,2,1)
a1.bar(x1,y1)
a1.set_title('A')
a2=fig.add_subplot(1,2,2)
a2.bar(x2,y2)
a2.set_title('B')
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y1=np.sin(x)
y2=np.cos(x)
y3=np.sin(-x)
y4=np.cos(-x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.show()
```
