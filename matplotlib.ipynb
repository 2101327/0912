```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
t=np.arange(0,2,0.01)
s=1+np.cos(2*np.pi*t)
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='x',ylabel='y',title='y=cos(2Ï€x)+1')
ax.grid()
plt.show()
```
